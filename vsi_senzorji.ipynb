```python
def find_adjacent_squares(sensor, map, osnova):
    # Ensure that both GeoDataFrames are using the same CRS
    if sensor.crs != map.crs:
        sensor = sensor.to_crs(map.crs)

    # Assuming sensor_square_gdf contains only one sensor square
    sensor_geometry = sensor.iloc[0].geometry

    # Find all squares that touch the sensor square's geometry
    # "touches" returns True for geometries that share at least one point (but no interior points)
    adjacent_squares = map[map.touches(sensor_geometry)]

    filtrirani = adjacent_squares[adjacent_squares["mean"] < osnova]

    return filtrirani
```

```python
def predikcija_senzorja(senzor):
    sensor_data = {
    'sensor_id': senzor,
    'longitude': [slovar_pretokov[senzor][3]],  # Example longitude of the sensor in Slovenia
    'latitude': [slovar_pretokov[senzor][4]]    # Example latitude of the sensor in Slovenia
    }
    sensor_gdf = gpd.GeoDataFrame(sensor_data, geometry=gpd.points_from_xy(sensor_data['longitude'], sensor_data['latitude']))
    sensor_gdf.crs = "EPSG:4326"

    if sensor_gdf.crs != map_gdf.crs:
        sensor_gdf = sensor_gdf.to_crs(map_gdf.crs)
    squares_with_sensor = gpd.sjoin(map_gdf, sensor_gdf, how="inner", op='contains')

    # Assuming you want the first sensor and its corresponding square
    sensor_square = squares_with_sensor.iloc[0]  # This returns a Series

    osnova1 = sensor_square["mean"]

    adjacent_squares_gdf = find_adjacent_squares(squares_with_sensor, map_gdf, osnova1)
    combined_gdf = squares_with_sensor
    combined_gdf = gpd.GeoDataFrame(pd.concat([combined_gdf, adjacent_squares_gdf], ignore_index=True))
    all = combined_gdf.copy()

    n = osnova1
    for _ in range(6):
        n -= 3
        for index, square in combined_gdf.iterrows():
        # Convert the current row 'square' to a GeoDataFrame
            single_square_gdf = gpd.GeoDataFrame([square], geometry='geometry', crs=combined_gdf.crs)
            temporary = find_adjacent_squares(single_square_gdf, map_gdf, n)
    
            all = gpd.GeoDataFrame(pd.concat([all, temporary], ignore_index=True))
            all = all.drop_duplicates(subset='geometry')

        combined_gdf = all.copy()
    all = all.drop_duplicates(subset='geometry')
    return all
```
